```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
import cv2
import random
import screeninfo
```

```python
face_cascade = cv2.CascadeClassifier('haarcascades/haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('haarcascades/haarcascade_eye.xml')
```

```python

 
screen = screeninfo.get_monitors()[0]

cap = cv2.VideoCapture(0)
 
img_counter = 0    
 
canvas = np.zeros((1080, 1920, 3), dtype = "uint8")
rnd_x = random.randint(1,1920)
rnd_y = random.randint(1,1080)
cv2.circle(canvas, (450, 303), 10, (0, 0, 255), 3)  

cv2.namedWindow('dark', cv2.WND_PROP_FULLSCREEN)
cv2.moveWindow('dark', screen.x - 1, screen.y - 1)
cv2.setWindowProperty('dark', cv2.WND_PROP_FULLSCREEN,
                      cv2.WINDOW_FULLSCREEN)

cv2.imshow('dark', canvas)
    
while(True):
    #leemos un frame y lo guardamos
    ret, img = cap.read()
 
    #convertimos la imagen a blanco y negro
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
 
    #buscamos las coordenadas de los rostros (si los hay) y
    #guardamos su posicion
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
 
    #Dibujamos un rectangulo en las coordenadas de cada rostro
    for (x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_gray)
        try:
            ex1,ey1,ew1,eh1 = eyes[0][0], eyes[0][1], eyes[0][2], eyes[0][3]
            ex2,ey2,ew2,eh2 = eyes[1][0], eyes[1][1], eyes[1][2], eyes[1][3]
            #cv2.rectangle(roi_color,(ex1,ey1),(ex2+ew2,ey2+eh2),(255,0,0),2)
            roi_eye_color = roi_color[ey1:ey2+eh2, ex1:ex2+ew2]
            #cv2.imshow('eyes',roi_eye_color)
        except:
            pass
 
    #Mostramos la imagen

    
    #cv2.imshow('img',img)
    
    
    wk = cv2.waitKey(1)
    

    
    if wk == ord('s'):
        canvas = np.zeros((1080, 1920, 3), dtype = "uint8")    
           
        cv2.circle(canvas, (rnd_x, rnd_y), 10, (0, 0, 255), 3) 
        
        font = cv2.FONT_HERSHEY_SIMPLEX 
        
        cv2.putText(canvas, 'Imagenes guardadas: '+str(img_counter+1), (100, 100), font, 0.8, (255, 255, 0), 2, cv2.LINE_AA)
        
        cv2.imshow('dark', canvas)
               
        if roi_eye_color.shape[1]/(roi_eye_color.shape[0]+0.0001)>2:
            try:
                cv2.imwrite('G:\Mi unidad\Personal\FASTAI DATA\GIT\Playground\Cascade\imgs\eye{}-{}-{}.png'.format(img_counter,
                                                                                                                   rnd_x,
                                                                                                                   rnd_y), roi_eye_color)
                img_counter += 1
                print('Image saved')
                rnd_x = random.randint(1,1920)
                rnd_y = random.randint(1,1080)



            except:
                print('Eyes not detected')
        else:
            pass

    else:
        pass  
              
    if wk == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
